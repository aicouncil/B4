```python
#Login activity => allow user to provide loginid and password for 3 times

for i in range(0,3):
  userid = input("Enter your user id ")
  userpass = input("Enter the password ")

  if userid == 'admin' and userpass == 'password':
    print("You are logged in")
    break
  else:
    print("Incorrrect credentials")
```

```python
#take a number as user input
#create a multiplication table from 1 to 10 for the number entered by user

number= int(input("Enter a number: "))

i = 1

while i <= 10:
    print(f"{number} × {i} = {number * i}")
    i += 1
```

```python
#create a function which can check a number is even or not

def chk_even_odd(number):
  if number % 2 == 0:
    print ("even")
  else:
    print ("odd")
```

```python
chk_even_odd(9)
```

```python
chk_even_odd(8)
```

```python
def compare(a,b):
  if a > b:
    print("First number is greater compared to Second")
  else:
    print("Second number is greater compared to First")
```

```python
compare(9,4)
```

```python
compare(9,14)
```
